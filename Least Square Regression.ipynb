```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("C:/Users/Micaela Campher/Desktop/weightdataset.csv") 
df.tail()
```

```python
scaling=StandardScaler()
```

```python
weight=scaling.fit_transform(df[['Height']])
height=scaling.fit_transform(df[['Weight']])
plt.plot(weight, height,'.')
plt.show()
```

```python
x = weight
y = height 

x_mean= np.mean(x)
y_mean= np.mean(y)

num=0 
den=0 
```

```python
for i in range(len(x)):
    num += (x[i]-x_mean)*(y[i]-y_mean)
    den += (y[i]-x_mean**2)
    
m= num/den
c=y_mean-m*x_mean

print(m,c)
```

```python
y_pred = m* x + c
y_pred=scaling.fit_transform(y_pred)
plt.plot(x,y,'.')
plt.plot([min(x),max(y)], [min(y_pred), max(y_pred)])
plt.title("Height vs Weight")
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```
